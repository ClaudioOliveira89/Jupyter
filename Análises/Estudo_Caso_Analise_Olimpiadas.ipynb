```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
países = ['EUA', 'China', 'Brasil', 'Argentina', 'Alemanha']
```

```python
bronze = np.array([38,17,26,19,15])
```

```python
prata = np.array([12,56,78,12,34])
```

```python
ouro = np.array([24,56,12,57,89])
```

```python
ind = np.arange(len(países))
```

```python
ind
```

```python
plt.bar(ind, ouro, label='Ouro', color='gold', bottom=prata+bronze)
plt.bar(ind, prata, label='Prata', color='silver', bottom=bronze)
plt.bar(ind, bronze, label='Bronze', color='#CD853F')
plt.xticks(ind, países)
plt.ylabel("Medalhas")
plt.xlabel("Países")
plt.legend()
plt.title("Medalhas - Olimpíadas 2018")
plt.show()
```
