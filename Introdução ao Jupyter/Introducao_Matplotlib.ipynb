```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,20)
```

```python
x
```

```python
Y = x ** 2 
```

```python
Y
```

```python
plt.plot(x,Y)
plt.xlabel('Variável do eixo x')
plt.ylabel('Variável do eixo y')
plt.title("Gráfico 1")
plt.show()
```

```python
linguagens = ('Python', 'C++', 'Java', 'Pearl', 'Scala', 'Lips')
```

```python
x = np.arange(len(linguagens))
```

```python
x
```

```python
y = [10,8,6,7,5,3]
```

```python
plt.bar(x,y, align ='center')
```

```python
plt.bar(x, y, align ='center')
plt.xticks(x, linguagens)
plt.ylabel("Taxa de uso")
plt.title("Uso de linguagens de programação")
```

```python
labels = 'Android', 'Blackberry', 'Windows Phone', 'iOS'
```

```python
vendas = [ 70, 5, 10, 25]
```

```python
cores = ['lightyellow', 'lightgreen', 'lightcoral', 'lightskyblue']
```

```python
explode = (0.1, 0, 0, 0)
```

```python
plt.pie(vendas, explode=explode, labels=labels, colors=cores, autopct='%1.1f%%', shadow=True, startangle=0)
plt.axis("equal")
plt.show()
```
